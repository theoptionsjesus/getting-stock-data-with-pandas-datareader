```python
import datetime as dt
import pandas as pd
from pandas_datareader import data as pdr 

```

```python
end = dt.datetime.now()
start = dt.datetime(2010,1,1)
print ("start:", start)
print ("end:", end)




```

```python
stocklist = ["^VIX","^GSPC","^IXIC"] 
print (stocklist)







```

```python
df = pdr.get_data_yahoo(stocklist, start, end)
print (df.head(20))








```

```python
close = df.Close
print (close.head(20))









```

```python
print(close.describe(percentiles = [0.1, 0.5, 0.9]))





```
